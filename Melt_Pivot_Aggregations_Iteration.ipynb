```python
import pandas as pd 
import numpy as np 
```

```python
diabetic_df = pd.read_csv('diabetic_data.csv')
diabetic_df.head()
```

```python
#Melt
diabetic_melt = pd.melt(diabetic_df, id_vars=['patient_nbr'], value_vars=['diabetesMed'], var_name = 'Diabetes Medication') 
#melting the patient number, as this identifies each individual patient, alongside the diabetesMed column to understand how the particular patient is related to the diabetesMed column 
```

```python
print(diabetic_melt)
```

```python
#Pivot
diabetic_pivot = diabetic_df.pivot(values = 'number_diagnoses', index = 'encounter_id', columns = 'readmitted')
#applying pivot table method to understand how the number of diagnoses is related to readmission for individual encounters
print(diabetic_pivot)
```

```python
#Aggregation
diabetic_agg = diabetic_df.aggregate({'time_in_hospital':['mean','min','max'], 'num_lab_procedures':['mean','min','max'], 'num_medications': ['mean','min','max']})
#using aggregation to understand how the time spent in the hospital may correlate to the number of lab procedures done and medications the patients takes 
#utilizing the mean, min and max of the variables 
print(diabetic_agg)
```

```python
#Itteration
for index, row in diabetic_df.iterrows(): #using iterrow to itterate through the 'race' column in the dataset - this repeatedly executes each race in the dataset
    print(row['race'])
```

```python
#Group By
diabetes_groupby = diabetic_df.groupby('gender') #this step actually groups the chosen column
diabetes_groupby.describe() #this step aggregates, in this case using describe method, which outputs various descriptions as seen in the table output
```
